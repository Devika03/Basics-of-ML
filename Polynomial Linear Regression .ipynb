```python
import pandas as pd
import numpy as np 

import matplotlib.pyplot as plt 
import seaborn as sns
```

# Loading Data

```python
data = pd.read_csv("Position_Salaries.csv")
```

```python
data
```

# Splitting of the Data

```python
x= data.iloc[: , 1:-1].values #Input parameter - All rows and Level column .
y= data.iloc[: , -1].values#Output parameter - All rows and Salary column .
```

```python
x
```

```python
y
```

# Linear Regression Model 

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

# Visualize the LR Model

```python
plt.scatter(x , y , color = "red")
plt.plot(x , lin_reg.predict(x) , color = "blue")
plt.xlabel("Input")
plt.ylabel("Prediction")
plt.show()
```

# Polynomial Linear Regression

```python
from sklearn.preprocessing import PolynomialFeatures 

poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x) #1. Fitting the data , 2. As degree = 2  so It is doing x ^ 0 , x^1 , x^ 2. 
```

```python
x_poly
```

```python
poly_linear_reg = LinearRegression()
poly_linear_reg.fit(x_poly , y )
```

# Visualize Polynomial Linear Regression

```python
#Degree = 2 
plt.scatter(x , y , color = "red")
plt.plot(x , poly_linear_reg.predict(poly_reg.fit_transform(x)) , color = "blue")
plt.xlabel("Input")
plt.ylabel("Prediction")
plt.show()
```

```python
#After increasing the degree to 3 
plt.scatter(x , y , color = "red")
plt.plot(x , poly_linear_reg.predict(poly_reg.fit_transform(x)) , color = "blue")
plt.xlabel("Input")
plt.ylabel("Prediction")
plt.show()
```

```python
#It fits better 
#Higher the degree , Better it fits. 
```

```python
#Degree = 4 
plt.scatter(x , y , color = "red")
plt.plot(x , poly_linear_reg.predict(poly_reg.fit_transform(x)) , color = "blue")
plt.xlabel("Input")
plt.ylabel("Prediction")
plt.show()
```

```python
#Extremely high degrees are not used , then your model tends to mimic your data. 
```
