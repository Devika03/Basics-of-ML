# Data Processing 

```python
data = data.dropna()
print('Shape of the data : ' , data.shape)
```

# Splitting data

```python
#training input and output
train_input = np.array(data.x[0:500]).reshape(500,1) #We are taking 0 to 499 x_train values. 
train_output = np.array(data.y[0:500]).reshape(500,1)

#Validation data input and output 
test_input = np.array(data.x[500:700]).reshape(199,1)
test_output = np.array(data.y[500:700]).reshape(199,1)
```

```python
np.array(data.x[0:500])
```

```python
np.array(data.x[0:500]).reshape(500,1)
```

```python
print(f"Training input data shape =  {train_input.shape }")
print(f"Training output data shape = {train_output.shape }")
print(f"Testing  input  data shape = {test_input.shape }")
print(f"Testing  output data shape = {test_input.shape }")
```

# Linear Regression

```python
#Training Model 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()#Initialized 
```

```python
lr_model.fit(train_input , train_output) #fit() will perform all steps of LR 
```

```python
lr_model.coef_ #It will return you the optimal m-value 
```

```python
lr_model.intercept_ #It will return the optimal c-value 
```

# Prediction

```python
test_predictions = lr_model.predict(test_input) #It will take m and c and apply on test_input 
```

```python
plt.plot(test_input , test_predictions , "+" , color = "red")
plt.plot(test_input , test_output , "+" , color = "green")
plt.title("Performance Testing")
plt.xlabel("Input")
plt.ylabel("Output / Predictions")
plt.show()
```

# Cost

```python
from sklearn.metrics import mean_squared_error
```

```python
cost = mean_squared_error(test_output , test_predictions)
```
