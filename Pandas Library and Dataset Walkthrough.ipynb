# Total number of apps in each category

```python
categories = {}

for i in df['Category'].unique():
    ct=0
    for j in df['Category']:
        if (i ==j ):
            ct += 1
    categories[i] = ct 
print(categories)
print(categories['ART_AND_DESIGN'])
```

# Total no. of apps in each Type 

```python
types = {}
for name in df['Type'].unique():
    ct=0
    for i in df['Type']:
        if ( i == name ):
            ct += 1
    types[name]= ct
print(types)
```

# Total no. of apps in each Content rating 

```python
content_rating = {}
for name in df['Content Rating'].unique():
    ct = 0
    for i in df['Content Rating']:
        if(i==name):
            ct+=1
    content_rating[name]=ct
print(content_rating)
```

# Method 2. Imputer .

```python
import sklearn 
from sklearn.impute import SimpleImputer

import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

```python
impute = SimpleImputer(missing_values = np.nan , strategy = 'mean') #We will take mean rating
```

```python
df.iloc[: , 2:3 ].values # To select Rating
```

```python
impute.fit(df.iloc[: , 2:3 ].values)
```

```python
df.iloc[: , 2:3 ] = impute.transform(df.iloc[: , 2:3 ].values)
```

```python
df.head() #Un-edited
```

```python
df #Rating is edited 
```

```python
df.isnull().sum() #Previous Rating = 1474 ; Now Rating = 0 
```
